```python
f = open('bioDBnet_db2db_221016050334_2063528135.txt', 'r')
lines = f.readlines()

ensembl_items = list()
for line in lines[1:]:
    items = line.split('\t')[1]
    if items == '-': continue
    items = items.split(';')
    for item in items:
        item = item.strip()
        try:
            if item != '': ensembl_items.append(int(item.replace('ENSP','')))
        except:
            pass

len(ensembl_items)
```

```python
import numpy as np
import random
random.seed(42)

vectors = list()
ids = list()
labels = list()

f = open("./p10q5.emb", 'r')
lines = f.readlines()

X = []
y = []

for line in lines[1:]:
    line = list(map(float, line.split()))
    key = int(line[0])
    value = np.array(line[1:])
    
    vectors.append(value)
    ids.append(key)
    labels.append(0)
    
    if key in ensembl_items:
        labels[-1] = 1
        X.append(value)
        y.append(1)
    else:
        if random.random() > 0.9:
            X.append(value)
            y.append(0)
vectors = np.array(vectors)
X = np.array(X)
```

```python
print(f'# of targets: {y.count(1)}\n# of non-targets: {y.count(0)}\n# of dataset: {len(y)}')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier

X_trainvalid, X_outer, y_trainvalid, y_outer = train_test_split(X, y, test_size=0.1, random_state=42)

xgmodel = XGBClassifier()

xgb_param_grid={
    'n_estimators' : [100,300, 500, 1000, 2000, 3000],
    'learning_rate' : [0.01,0.05,0.1,0.15, 0.2, 0.3],
    'max_depth' : [2, 3, 5, 7, 10, 20],
    'gamma' : [0,1,2,3],
    'colsample_bytree' : [0.8,0.9],
}

grid=GridSearchCV(xgmodel, param_grid = xgb_param_grid, scoring="f1_macro", n_jobs=-1, verbose = 2, cv=5)
grid.fit(X_trainvalid, y_trainvalid)
```

```python
print('# of outer dataset: ', len(X_outer), '\n', '# of train/valid dataset: ', len(X_trainvalid))
```

```python
import pandas as pd

print("best f1_macro : {0: .4f}".format(grid.best_score_))
print("best param : ",grid.best_params_)

result_df = pd.DataFrame(grid.cv_results_)
result_df.sort_values(by=['rank_test_score'],inplace=True)

result_df.to_csv('./grid_results.csv')
result_df[['params','mean_test_score','rank_test_score']].head(10)
```

```python
xgmodel = XGBClassifier(colsample_bytree=0.9, gamma=1, learning_rate=0.3, max_depth=5, n_estimators=100, random_state=42)
xgmodel.fit(X_trainvalid, y_trainvalid)

y_pred = xgmodel.predict_proba(X_outer)
y_pred = y_pred[:,1]
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# y_true is the true labels of the data
# y_scores is the predicted probabilities (or scores)
precision, recall, thresholds = precision_recall_curve(y_outer, y_pred)

# Plot the precision-recall curve
plt.rc('axes', labelsize=20) 
plt.rc('axes', titlesize=20)
plt.plot(recall, precision, 'b-',marker='o',markersize=6, label='Precision-Recall curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.axis([0, 1, 0, 1])
plt.show()
```

```python
plt.plot(thresholds, precision[:len(thresholds)], label = 'precision')
plt.plot(thresholds, recall[:len(thresholds)], label = 'recall')

plt.axvline(0.05, 0, 1, color='gray', linestyle='--', linewidth=2)
plt.text(0.06, 0.15, 'Threshold 0.05(Recall ~90%)')

plt.grid()
plt.legend()
plt.xlabel('Threshold')
plt.show()
```

```python
import math

ths = [0.80, 0.85, 0.90, 0.95, 0.99]
for r, t in reversed(list(zip(recall[:len(thresholds)], thresholds))):
    if ths[0] < r:
        print(f'{int(ths[0]*100)}% recall rate threshold: {round(t, 5):.5f}')
        del ths[0]
```
